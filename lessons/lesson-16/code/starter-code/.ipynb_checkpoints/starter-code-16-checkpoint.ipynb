```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('../dataset/rossmann.csv', skipinitialspace=True)

data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

store1_data = data[data.Store == 1]

store1_open_data = store1_data[store1_data.Open == 1]

store1_open_data[['Sales']].plot()
```

```python
# correlation of data points

print store1_data.Sales.autocorr(lag=1)
print store1_data.Sales.autocorr(lag=7)
```

```python
from pandas.tools.plotting import autocorrelation_plot

autocorrelation_plot(store1_data.Sales)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
print plot_acf(store1_data.Sales, lags=25)
```

```python
from statsmodels.tsa.arima_model import ARMA

store1_sales_data = store1_open_data[['Sales']].astype(float)
model = ARMA(store1_sales_data, (1,0)).fit()
model.summary()
```

Build and AR(1) model. Notice that the coefficient is not the same as the autocorrelation at lag 1. Therefore, the assumptions that I started out with to do that derivation for autocorrelation = coefficient at lag 1 are not true. Therefore, the series is not stationary.

```python
model.resid.plot()
```

```python
print plot_acf(model.resid, lags=50)
```

```python
model = ARMA(store1_sales_data, (1, 1)).fit()
model.summary()
```

```python
model.resid.plot()
```

```python
print plot_acf(model.resid, lags=50)
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(store1_sales_data, (1, 0, 1)).fit()
model.summary()
```

```python
model = ARIMA(store1_sales_data, (1, 1, 2)).fit()
model.summary()
```

```python
store1_sales_data.Sales.diff(1).autocorr(lag = 1)
```

```python
print model.plot_predict(1, 50) # (starting point, ending point)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax = store1_sales_data['2014'].plot(ax=ax)
fig = model.plot_predict(1, 200, ax=ax, plot_insample=False)
```

### Walmart Sales Data

For the independent practice, we will analyze the weekly sales data from Walmart over a two year period from 2010 to 2012.

The data is again separated by store and by department, but we will focus on analyzing one store for simplicity.

The data includes:

- Store - the store number
- Dept - the department number
- Date - the week
- Weekly_Sales -  sales for the given department in the given store
- IsHoliday - whether the week is a special holiday week


#### Loading the data and setting the DateTimeIndex

```python
data = pd.read_csv('../dataset/train.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data.head()
```

#### Filter the dataframe to Store 1 sales and aggregate over departments to compute the total sales per store.

```python
# TODO
store1_sales = data[data.Store == 1].Weekly_Sales.resample('W', 'sum') # aggregate over departments
print store1_sales.head()
```

#### Plot the rolling_mean for `Weekly_Sales`. What general trends do you observe?

```python
# TODO

pd.rolling_mean(store1_sales,window=3, center = True).plot()
```

#### Split the weekly sales data in a training and test set - using 75% of the data for training

```python
# TODO
L = len(store1_sales) # 143 weeks
train_data = store1_sales[0:int(0.75*L)]
test_data = store1_sales[int(0.75*L):L]
```

#### Create an AR(1) model on the training data and compute the mean absolute error of the predictions.

```python
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
```

```python
# TODO

mymodel = sm.tsa.ARIMA(train_data, (1,0,0)).fit()
```

```python
# dynamic = True: use forecasted values as well to forecast

test_predictions = mymodel.predict('2012-02-26', '2012-10-28', dynamic = True)
```

```python
mean_absolute_error(test_data, test_predictions) / np.mean(test_data)
```

#### Plot the residuals - where are their significant errors.

```python
# TODO

mymodel.resid.plot()
```

#### Compute and AR(2) model and an ARMA(2, 2) model - does this improve your mean absolute error on the held out set.

```python
# AR(2)

mymodel_2 = sm.tsa.ARIMA(train_data, (2,0,0)).fit()
test_predictions_2 = mymodel_2.predict('2012-02-26', '2012-10-28', dynamic = True)
mean_absolute_error(test_data, test_predictions_2)
```

```python
# TODO

mymodel_22 = sm.tsa.ARIMA(train_data, (2,0,2)).fit()
test_predictions_22 = mymodel_22.predict('2012-02-26', '2012-10-28', dynamic = True)
mean_absolute_error(test_data, test_predictions_22)
```
